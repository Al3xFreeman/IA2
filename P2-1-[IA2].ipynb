- Cargamos el dicionario de palabras 

```python
import numpy as np
import numpy.ma as ma

def write_terms (feature_names, data, vector_data, index):
    '''
    Escribe los términos presentes en un mensaje representado como bolsa de palabras.
    
    - feature_names: terminos usados para vectorizar
    - data: lista de mensajes original (si data==None no se muestra el mensaje original)
    - vector_data: matriz (dispersa) de mensaje vectorizados
    - index: posición del mensaje a mostrar
    '''
    # máscara para seleccionar sólo el mensaje en posición index
    mask=vector_data[index,:]>0
    
    # términos que aparecen en ese mensaje vectorizado
    terminos = ma.array(feature_names, mask = ~(mask[0].toarray()))
    
    # mostrar mensaje original
    if data:
        print('Mensaje', index, ':', data[index])
    
    # mostrar términos que aparecen en el mensaje vectorizado
    print('Mensaje', index, 'vectorizado:', terminos.compressed(),'\n')
```

```python
from sklearn import tree
```

        - Bolsa de palabras binarias con n-grama (1,1): Usaremos un "arbol de decision" como clasificador. Vamos a ajustar los valores de  max_depth, min_samples_leaf o max_leaf_nodes para maximizar la tasa de aciertos. Para ello, vamos ir modificando los valores de los pametros. Haremos un bucle doble que compruebe cada una de las combinaciones para todos los posibles valores de nuestro rango. Paralelamente dibujaremos las graficas para ajustar los valores de max_depth y min_samples_leaf.

```python
import matplotlib.pyplot as plt

train_accuracy = []
test_accuracy = []
best = 0
bestD = 0
bestM = 0

max_depths = range(1, 30)
min_samples_leafs = range(1, 10)

for msl in min_samples_leafs:
    for md in max_depths:
        clf = tree.DecisionTreeClassifier(max_depth=md, min_samples_split=2, min_samples_leaf=msl)
        clf.fit(train_vector_data, train_sentiment)
        tree_train_predictions = clf.predict(train_vector_data)
        tree_test_predictions = clf.predict(test_vector_data)
        train_accuracy.append(np.mean(tree_train_predictions == train_sentiment))
        test_accuracy.append(np.mean(tree_test_predictions == test_sentiment))
        if (np.mean(tree_test_predictions == test_sentiment) > best):
            best = np.mean(tree_test_predictions == test_sentiment)
            bestD = md
            bestM = msl
    # Draw lines
    plt.plot(max_depths, train_accuracy, color="r",  label="Training")
    plt.plot(max_depths, test_accuracy, color="g", label="Test")
    # Create plot
    print("min_samples_leafs: " + str(msl))
    plt.title("Curva de aprendizaje")
    plt.xlabel("Depth"), plt.ylabel("Accuracy Score"), plt.legend(loc="best")
    plt.tight_layout()
    plt.show()
    print(train_accuracy)
    print(test_accuracy)
    train_accuracy=[]
    test_accuracy=[]
    
print()
print("RESULTADOS:")
print(best)
print(bestD)
print(bestM)
```

        Gracias a estas curvas de aprendizaje podemos ver la tendencia de la curva para cada uno de los valores. Sin embargo, el bucle acaba mostrando el mejor primer mejor resultado del test y que valores de max_depth y min_samples_leaf.
