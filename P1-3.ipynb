La correlación con el GDP está claramente presente en aquellos paises que están más desarrollados, que que es inversamente proporcional a métricas como la mortalidad infantil, tienen menor número de nacimientos o no dependen en una economñía basada en la agricultura. También una métrica que indica un alto GDP es el número de móviles por 1000 habitantes, ya que nos indica que en dicho pais habrá un mayor consumo de bienes que no son de primera necesidad. 

Otros elementos como el porcentaje de gente con educación parecen ser también bastante relevantes.
