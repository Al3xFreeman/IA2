### 2) 
Consideramos que es necesario re-escalar las variables ya que en concreto la variable 'Assault' tiene valores muy superiores al resto. Esto es importante tenerlo en cuenta ya que si no re-escalasemos las variables, la variable 'Assault' tendrá un peso mayor al resto sin que nosotros se lo hayamos dado. Para ello vamos a normalizar los datos de dos maneras distintas (MinMaxEscaler y StandarScaler) y luego comparar sus resultados.

```python
# Limpiamos el dataset.
estados = data['Unnamed: 0']
seguridad = data['Safety']
data2 = data.drop(['Unnamed: 0', 'Safety'], axis=1)
```

```python
from sklearn import preprocessing
```

```python
# Usamos una normalización con MinMaxScaler.
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(data2.values)
scaled_Data_MinMax = pd.DataFrame(x_scaled, columns=data2.keys())
scaled_Data_MinMax['Name states'] = estados
scaled_Data_MinMax['Safety'] = seguridad
scaled_Data_MinMax
```

```python
# Usamos una normalización con MinMaxScaler.
standar_scaler = preprocessing.StandardScaler()
y_scaled = standar_scaler.fit_transform(data2.values)
scaled_Data_Standar = pd.DataFrame(y_scaled, columns=data2.keys())
scaled_Data_Standar['Name states'] = estados
scaled_Data_Standar['Safety'] = seguridad
scaled_Data_Standar
```

Como se puede observar en las dos tablas superiores, hemos hecho dos tipos de escalado de los datos:
Por un lado hemos usado el escalado minMax que deja todos los valores compredidos entre 0 y 1. Por el otro, hemos usado un escalado estandar que basicamente deja los varoles de las variables entre -1 y 1.
A partir de ahora haremos los siguientes puntos en base a estos dos tipos de reescalado.

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import davies_bouldin_score
from sklearn.datasets.samples_generator import make_blobs
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
# Eliminamos las variables que no vamos a tener en cuenta.
cluster_minMax_data = scaled_Data_MinMax.drop(['Safety', 'UrbanPop', 'Name states'], axis=1)
cluster_estandar_data = scaled_Data_Standar.drop(['Safety', 'UrbanPop', 'Name states'], axis=1)
```

```python
cluster_minMax_data
```

```python
cluster_estandar_data
```

```python
def plot_clusters(X, labels=None, centers=None):
    """ Función auxiliar para pintar los puntos de los clusters y, optativamente, sus centros.
    :param X: array de puntos de dos dimensiones (array de array de enteros)
    :param labels: cluster al que pertenece cada punto (array de enteros)
    :param centers: coordenadas de los centroides de los clusters (array de array de enteros)
    """
    
    colors = ['r','g','b','c','y','k','w','m',]
    fig = plt.figure(figsize=(8,8))
    
    # Los dos primeros parámetros de la función scatter son las coordenadas de los puntos,
    # el parámetro 'c' indica la clase asignada de cada punto, y 'cmap' se usa para colorear 
    # las distintas clases
    plt.scatter(X[:,0], X[:,1], c=labels, cmap=ListedColormap(colors))
    
    # Pintar los centroides de los clusters
    if centers is not None:
        plt.scatter(centers[:,0], centers[:,1], marker="x", color='k', s=150, linewidths = 5, zorder=10)
        
    plt.show()
```

Aplicamos k-medias

```python
# Min-Max
K_MAX = 8
davies_boulding = np.zeros(K_MAX-2)
for k in range(2, K_MAX): 
    km = KMeans(init='random', n_clusters=k, random_state=333)
    km.fit(cluster_minMax_data)
    plot_clusters(cluster_minMax_data.values, km.labels_, km.cluster_centers_)
    
    davies_boulding[k-2] = davies_bouldin_score(cluster_minMax_data, km.labels_) 
```

```python
# Estandar
davies_boulding_s = np.zeros(K_MAX-2)
for k in range(2, K_MAX): 
    kmS = KMeans(init='random', n_clusters=k, random_state=333)
    kmS.fit(cluster_estandar_data)
    plot_clusters(cluster_estandar_data.values, kmS.labels_, kmS.cluster_centers_)
    
    davies_boulding_s[k-2] = davies_bouldin_score(cluster_estandar_data, kmS.labels_) 
```

Ahora pasamos a dibujar una grafica y buscar el valor mas pequeño que corresponde al mejor numero de clusters a usar.

```python
# Min-Max
plt.plot(range(2, K_MAX), davies_boulding)
plt.title('Grafica de calidad de los cluster.')
```

```python
# Estandar
plt.plot(range(2, K_MAX), davies_boulding_s)
plt.title('Grafica de calidad de los cluster.')
```

Como se puede observar en ambos casos, el mejor numero de clusters a elegir es 2.

Aplicamos jerarquico

```python
# Min-Max
K_MAX = 8
davies_boulding_m = np.zeros(K_MAX-2)
for k in range(2, K_MAX): 
    ag = AgglomerativeClustering(n_clusters=k, linkage='ward', affinity='euclidean')
    ag.fit(cluster_minMax_data)
    plot_clusters(cluster_minMax_data.values, ag.labels_)
    
    davies_boulding_m[k-2] = davies_bouldin_score(cluster_minMax_data, ag.labels_) 
```

```python
# EStandar
K_MAX = 8
davies_boulding_e = np.zeros(K_MAX-2)
for k in range(2, K_MAX): 
    agS = AgglomerativeClustering(n_clusters=k, linkage='ward', affinity='euclidean')
    agS.fit(cluster_estandar_data)
    plot_clusters(cluster_estandar_data.values, agS.labels_)
    
    davies_boulding_e[k-2] = davies_bouldin_score(cluster_estandar_data, agS.labels_) 
```

```python
# Min - Max
plt.plot(range(2, K_MAX), davies_boulding_m)
plt.title('Grafica de calidad de los cluster.')
```

De nuevo la repressentacion de la grafica nos facilita el mejor valor de K, K = 2.

Para concluir este apartado diremos que nuestra eleccion del numero de clusters es de 2. Ya que nuestra interpretacion del problema y de los datos coincide en separar en dos grupos: Unos que podriamos considerar peligrosos y otros mas seguros. Por ultimo vamos a dibujar un dendograma para ilustrar como actua el algoritmo jerarquico.

```python
from scipy.cluster.hierarchy import dendrogram

def plot_dendrogram(model, **kwargs):

    # Children of hierarchical clustering
    children = model.children_

    # Distances between each pair of children
    # Since we don't have this information, we can use a uniform one for plotting
    distance = np.arange(children.shape[0])

    # The number of observations contained in each cluster level
    no_of_observations = np.arange(2, children.shape[0]+2)

    # Create linkage matrix and then plot the dendrogram
    linkage_matrix = np.column_stack([children, distance, no_of_observations]).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)
    

# Dibujar los primeros 5 niveles del árbol
levels = 5
plot_dendrogram(ag1, p=levels, truncate_mode='level')

# Árbol completo
#plot_dendrogram(ag)
```
