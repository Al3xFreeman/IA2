- Cargamos el dicionario de palabras 

```python
# Pasamos el fichero a una lista (una línea por item)
with open('words.txt') as f:
    dictionary = f.read().splitlines()
```

```python
import numpy as np
import numpy.ma as ma

def write_terms (feature_names, data, vector_data, index):
    '''
    Escribe los términos presentes en un mensaje representado como bolsa de palabras.
    
    - feature_names: terminos usados para vectorizar
    - data: lista de mensajes original (si data==None no se muestra el mensaje original)
    - vector_data: matriz (dispersa) de mensaje vectorizados
    - index: posición del mensaje a mostrar
    '''
    # máscara para seleccionar sólo el mensaje en posición index
    mask=vector_data[index,:]>0
    
    # términos que aparecen en ese mensaje vectorizado
    terminos = ma.array(feature_names, mask = ~(mask[0].toarray()))
    
    # mostrar mensaje original
    if data:
        print('Mensaje', index, ':', data[index])
    
    # mostrar términos que aparecen en el mensaje vectorizado
    print('Mensaje', index, 'vectorizado:', terminos.compressed(),'\n')
```

- Creamos bolsa de palabras binaria con usando un monograma (n-grama (1,1)):

```python
vectorizer = CountVectorizer(binary= True, vocabulary=dictionary, stop_words='english', ngram_range = (1,1))
```

```python
feature_names = vectorizer.get_feature_names()
```

```python
write_terms(feature_names, None, train_vector_data, 100)
```

    -  Arbol de decisión:

```python
from sklearn import tree
```

        - Bolsa de palabras binarias con n-grama (1,1): Usaremos un "arbol de decision" como clasificador. Vamos a ajustar los valores de  max_depth, min_samples_leaf o max_leaf_nodes para maximizar la tasa de aciertos. Para ello, vamos ir modificando los valores de los pametros. Haremos un bucle doble que compruebe cada una de las combinaciones para todos los posibles valores de nuestro rango. Paralelamente dibujaremos las graficas para ajustar los valores de max_depth y min_samples_leaf.

```python
from sklearn.model_selection import cross_validate
import matplotlib.pyplot as plt

train_accuracy = []
test_accuracy = []

max_depths = range(1, 20)
min_samples_leafs = range(1, 6)

data_text_vectorizer = vectorizer.fit_transform(data.text)

for msl in min_samples_leafs:
    for md in max_depths:
        clf = tree.DecisionTreeClassifier(max_depth=md, min_samples_split=2, min_samples_leaf=msl)
        scores = cross_validate(clf, data_text_vectorizer, data.sentiment, scoring='accuracy', cv=10, return_train_score=True)
        train_accuracy.append(np.mean(scores['train_score']))
        test_accuracy.append(np.mean(scores['test_score']))
    # Draw lines
    plt.plot(max_depths, train_accuracy, color="r",  label="Training")
    plt.plot(max_depths, test_accuracy, color="g", label="Test")
    # Create plot
    print("min_samples_leafs: " + str(msl))
    plt.title("Curva de aprendizaje")
    plt.xlabel("Depth"), plt.ylabel("Accuracy Score"), plt.legend(loc="best")
    plt.tight_layout()
    plt.show()
    train_accuracy=[]
    test_accuracy=[]
```

        Gracias a estas curvas de aprendizajes de los clasificadores sabemos que a medida que crece la profundidad, es decir, el valor de max_depth es mayor mejora tanto el train como el test (aunque en puntos parezca que sufre de sobre-aprendizaje la tendencia es ascendente). Por lo tanto fijaremos el valor de max_depth al que viene por defecto (lo extiende hasta el final). Por otro lado, podemos ver que al modificar el valor de min_samples_leaf la diferencia entre los clasificadores es muy pequeña. Aun asi, elegiremos el valor para min_samples_leaf = 1.

```python
# Creamos el clasificador con max_depth = por defecto y min_samples_leaf = 1
tree_classifier = tree.DecisionTreeClassifier(min_samples_leaf = 1)
tree_classifier.fit(train_vector_data, train_sentiment)

tree_train_predictions = tree_classifier.predict(train_vector_data)
tree_test_predictions = tree_classifier.predict(test_vector_data)

print("Árbol, porcentaje de aciertos en entrenamiento:", np.mean(tree_train_predictions == train_sentiment))
print("Árbol, porcentaje de aciertos en test:", np.mean(tree_test_predictions == test_sentiment))
```

        Como se puede ver los resultados de aciertos estan por encima del 50% y aunque sufre de sobre-aprendizaje segun el estudio anterior son los mejores valores. Tiene sentido que sea mejor el valor de min_samples_leaf = 1 ya que fuerza que como minimo se necesita una muestra en un nodo hoja. Si este valor fuera mayor podria haber nodos hoja que quizas no sean semejantes, bajando asi el porcentaje de aciertos tanto en test como en train.
